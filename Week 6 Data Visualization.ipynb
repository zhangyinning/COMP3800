```python
# import libraries and load the DataFrame
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

df = pd.read_csv('Sales.csv')
```

```python
print(df)
```

# Bar Plot

```python
# A simple bar plot
plt.bar(df.iloc[:, 0], df.iloc[:, 6] )
plt.show()
```

```python
# Add decorations to the chart. 
plt.bar(df.iloc[:, 0], df.iloc[:, 6] ) #data for the plot

plt.title('Sales Records')
plt.xlabel('Sales Representatives')
plt.ylabel('No# of Products')
plt.legend(['Total in a week'], loc = "upper right")

plt.yticks(np.arange(0, 20, 5))  # the numbers on the y axis

plt.show()
```

```python
# horizontal bar chart
plt.barh(df.iloc[:, 0], df.iloc[:, 6] ) #data for the plot

plt.title('Sales Records')
plt.xlabel('No# of Products')
plt.ylabel('Sales Representatives')
plt.legend(['Total in a week'], loc = "upper right")

plt.xticks(np.arange(0, 20, 5))  # the numbers on the y axis

plt.show()
```

```python
# Grouped Bar Chart

x_axis = np.arange(4)
width = 0.15
# the first parameter sets the positions on the x axis. 
# the second parameter sets the heights of the bars.
# the third parameter is the width of the bars.
plt.bar(x_axis, df.iloc[:, 1], width = width)   
# In grouped bar chart, the x-axis of the second group has to be moved right to the first group, x_axis+width. 
plt.bar(x_axis+width, df.iloc[:, 2], width = width)  
plt.bar(x_axis+2*width, df.iloc[:, 3], width = width)   
plt.bar(x_axis+3*width, df.iloc[:, 4], width = width)
plt.bar(x_axis+4*width, df.iloc[:, 5], width = width)

plt.title('Sales Records')
plt.xlabel('Sales Representatives')
plt.ylabel('No# of Products')
plt.legend(df.columns[1:6], loc = "upper right")

plt.xticks(x_axis, df.iloc[:, 0])
plt.yticks(np.arange(0, 10, 1))

plt.show()
```

```python
# Stacked Bar Chart
x_axis = np.arange(4)
width = 0.15

plt.bar(x_axis, df.iloc[:, 1], width = width) 
plt.bar(x_axis, df.iloc[:, 2], width = width, 
        bottom = df.iloc[:, 1])  # For stacked bar plot, the bottoms of the bars are raised
plt.bar(x_axis, df.iloc[:, 3], width = width, 
        bottom = df.iloc[:, 1]+df.iloc[:, 2])
plt.bar(x_axis, df.iloc[:, 4], width = width, 
        bottom = df.iloc[:, 1]+df.iloc[:, 2]+df.iloc[:, 3])
plt.bar(x_axis, df.iloc[:, 5], width = width, 
        bottom = df.iloc[:, 1]+df.iloc[:, 2]+df.iloc[:, 3]+df.iloc[:, 4])

plt.title('Sales Records')
plt.xlabel('Sales Representatives')
plt.ylabel('No# of Products')
plt.legend(df.columns[1:6], loc = "upper right")

plt.xticks(x_axis, df.iloc[:, 0])
plt.yticks(np.arange(0, 20, 2))

plt.show()
```

# Pie Chart

```python
plt.pie(df.iloc[:, 6],   # data source
        explode = [0.05, 0.05, 0.2, 0.05],  # this separates the pies from the center
        labels = df.iloc[:, 0],  # lables like Anna, Jason
        shadow = True, 
        autopct = '%.2f%%')  # This shows the pecentage numbers on the pie. 
plt.legend()
plt.show()
```

# Adding Subplots

```python
fig, axes = plt.subplots(2,3, sharex = True, sharey = True)
axes[0, 0].bar(df.iloc[:, 0], df.iloc[:, 1] )
axes[0, 0].set_title('Monday')

axes[0, 1].bar(df.iloc[:, 0], df.iloc[:, 2] )
axes[0, 1].set_title('Tuesday')

axes[0, 2].bar(df.iloc[:, 0], df.iloc[:, 3] )
axes[0, 2].set_title('Wednesday')

axes[1, 0].bar(df.iloc[:, 0], df.iloc[:, 4] )
axes[1, 0].set_title('Thursday')

fig.suptitle('Sales Records')

plt.show()
```

```python
fig, axes = plt.subplots(2,3, sharex = True, sharey = True)

z = 1
for x in range(2):
    for y in range(3):
        axes[x, y].bar(df.iloc[:, 0], df.iloc[:, z])
        z+=1

fig.suptitle('Sales Records, Monday - Friday')

plt.show()
```
