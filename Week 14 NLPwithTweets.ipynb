```python
query = "(from:taylorswift13) until:2022-11-06 since:2021-1-01"
tweets = []
limit = 5000


for tweet in sntwitter.TwitterSearchScraper(query).get_items():
    
    if len(tweets) == limit:
        break
    else:
        tweets.append([tweet.date, tweet.user.username, tweet.content, tweet.retweetCount])
        
df = pd.DataFrame(tweets, columns=['Date', 'User', 'Tweet', 'RetweetCount'])
print(df)
```

```python
# Put these sentences into a string
df1 = df['Tweet']

all_sentences = []

for sentence in df1:
    all_sentences.append(sentence)

all_sentences   # this returns a list with each sentence as an element
```

```python
sentences = " ".join(all_sentences)
sentences
```

# Keyword Search

```python
# Find all the @.. mentioned twitter accounts
import re

callouts = re.findall(r'@[a-zA-Z0-9_*]+', sentences)
print(callouts)
```

```python
# Count the frequencies of these mentioned names
import collections
freqs = collections.Counter(callouts)
print(freqs)
```

# Data Cleaning and Information Extraction

```python
# tokenizing
from nltk.tokenize import word_tokenize
tweets_tokens = word_tokenize(sentences)

tweets_tokens
```

```python
# remove punctuations
import re
tweets_tokens = [re.sub(r'[^A-Za-z0-9]+', '', x) for x in tweets_tokens]

words_no_punc = []

for word in tweets_tokens:
    if word != '':
        words_no_punc.append(word)
```

```python
# remove stopwords
import nltk
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english")) # this loads a set of the stopwords library

words_no_stopwords = []

for word in words_no_punc:
    if word.casefold() not in stop_words:   #.casefold(), similar to .lower().
        words_no_stopwords.append(word)

words_no_stopwords
```

```python
#stemming

from nltk.stem.snowball import SnowballStemmer

s_stemmer = SnowballStemmer(language='english')  # load the algorithm

sample_words = ['teach', 'teaching', 'taught', 'teacher', 'teaches']

new = []
for word in sample_words:
    new.append(s_stemmer.stem(word))

print(new)
```

```python
words_stemmed = []
for word in words_no_stopwords:
    words_stemmed.append(s_stemmer.stem(word))
    
words_stemmed
```

```python
# Lemmatization
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

lemmatizer.lemmatize("studies",pos='v')
```

# Sentiment Analysis

```python
# sample result
from textblob import TextBlob
print(TextBlob("I love COMP3800.").sentiment)
print(TextBlob("I do not love COMP3800.").sentiment)
print(TextBlob("I don't love COMP3800.").sentiment)
print(TextBlob("I very love COMP3800.").sentiment)
```

```python
# apply sentiment analysis on our project
def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity

# Create a function to get the polarity
def getPolarity(text):
    return  TextBlob(text).sentiment.polarity

df['Subjectivity'] = df['Tweet'].apply(getSubjectivity)
df['Polarity'] = df['Tweet'].apply(getPolarity)

df.head()
```
